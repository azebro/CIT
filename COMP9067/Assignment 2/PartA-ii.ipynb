Build a CNN ensemble containing a maximum of 10 base learners.

During the training process of each base learner you should use checkpointing so that you can capture the base learner model with the lowest validation loss. Clearly there are many types of ensembles that you could build and explore.

```python
#Load assignment dataset, as per provided code
def loadDataH5():
    with h5py.File('data/data1.h5','r') as hf: 
        trainX = np.array(hf.get('trainX')) 
        trainY = np.array(hf.get('trainY')) 
        valX = np.array(hf.get('valX')) 
        valY = np.array(hf.get('valY')) 
        print (trainX.shape,trainY.shape) 
        print (valX.shape,valY.shape)
    return trainX, trainY, valX, valY
trainX, trainY, testX, testY = loadDataH5()
```

```python

WIDTH = trainX.shape[2]
HEIGHT = trainX.shape[1]
DEPTH = trainX.shape[3]

#Assignment dataset has 17 different classes
CLASSES = 17
#Will run for 200 iterations
NUM_EPOCHS = 200
BATCH_SIZE = 64
```

```python
#Helper function to draw the output of model run
def visualiseOutput(history):
    
    print("\n")
    print("Min Val Loss: ", min(history.history['val_loss']))
    print("Max Val Accuracy: ", max(history.history['val_accuracy']))
    print("\n")
    
    
    plt.figure(figsize=(20,10))
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
   
    plt.title('Training Loss and Accuracy')
    plt.ylabel('Loss/Accuracy')
    plt.xlabel('Epoch #')
    plt.legend(['train loss', 'val loss', 'train acc', 'val acc'], loc='upper right')
    plt.show()
```

```python
'''
Main model execution function
It takes the following parameters:
    1. a model
    2. an augmentation generator
    3. optional model name to name the checkpoint files
It will fit the model and checkpoint the improved weights to the file
This function will produce 1 weights file per model, if multiple are required the fname varaible needs
to be swapped for the commented one
'''
def executeAugmentedModel(model, generator, modelName = "model1"):
    #fname ="outputs/" + modelName + "_weights.{epoch:02d}-{val_loss:.2f}-{val_accuracy:.3f}.hdf5"
    fname ="outputs/" + modelName + "_weights.hdf5"
    checkpoint = tf.keras.callbacks.ModelCheckpoint(fname, monitor="val_loss", mode="min", save_best_only=True, 
                                                    verbose=0)
    trainGenerator = generator.flow(trainX, trainY, batch_size= BATCH_SIZE)
    #Using SDG as advised in assignment
    opt = keras.optimizers.SGD(lr=0.01)
    model.compile(loss="sparse_categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
    h = model.fit(trainGenerator, 
                        validation_data=(testX, testY),  
                        epochs = NUM_EPOCHS, verbose=0, callbacks=[checkpoint])
    visualiseOutput(h)
    return min(h.history['val_loss']), max(h.history['val_accuracy'])
```

Helper functions for training base learners and ensemble execution:

```python
''' 
Base learner trainign function, it takes collection of models
and executes them, capturing stats
'''
def trainBaseLearners(models):
    learnerStats = []
    for model in models:
        name = model[1]
        print("Training model: {}. \n\n".format(name))
        l, a = executeAugmentedModel(model[0], trainDataGenerator, name)
        data = [name, l, a]
        learnerStats.append(data)
    return learnerStats
    
```

```python
'''
Helper function to load previously saved model weights
This function assumes that the models are saved in 'outputs' folder
and each file is named with the following convention:
    modelName_weights.hdf5
'''
def loadModels(models):
    loadedModels = []
    for model in models:
        name = model[1]
        path = "outputs/{}_weights.hdf5".format(name)
        m = model[0]
        m.load_weights(path)
        loadedModels.append(m)
   
    return loadedModels
```

```python
'''
Function calculating ensemble average scores
It takes list of models, and validation data and 
preforms prediction followed by averaging the results
'''
def ensembleModels(models, testX):
    # predcict each model, collect outputs of models in a list
    y = np.array([model.predict(testX) for model in models])
    #Average out
    result = np.argmax(np.average(y, axis=0), axis=1)
    return result
```

```python
#Helper function to visualise the ensamble
#Referenced from: Matplotlib documentation

def visualiseEnsemble(data):
    labels = data[:,0]
    # the label locations
    x = np.arange(len(labels)) 
    # the width of the bars
    width = 0.55  

    fig, ax = plt.subplots(figsize=(10,5))
    rects1 = ax.bar(x , data[:,2].astype(float), width)

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Accuracy')
    ax.set_title('Base Learners and Ensemble Accuracy')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)

    def autolabel(rects):
        #Attach a text label above each bar in *rects*, displaying its height.
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{:.3f}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(rects1)
    fig.tight_layout()
    plt.show()
```

###### Approach 2: Generate ensembles based on algorithm.

The algorithm will:
   1. randomly assign number of blocks
   2. randomly assign number of conv layers in each block
   3. Number of filters in each conv layer

```python
'''
Auto base learner generation factory function
It takes number of blocks and conv layers
'''
def modelFactory(blocks, nConv):
    model = keras.models.Sequential()
    inputShape = (HEIGHT, WIDTH, DEPTH)
    #Create initial block
    model.add(keras.layers.Conv2D(32, (3, 3), padding="same", input_shape=inputShape, activation='relu'))
    model.add(keras.layers.Conv2D(32, (3, 3), padding="same", activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
    #All other blocks are auto generated
    for i in range(blocks):
        for j in range(nConv):
            #Number of filters will increase with each added layer
            model.add(keras.layers.Conv2D((64 + 8 ** i), (3, 3), padding="same", activation='relu'))
        #Each block will finish with max pooling layer
        model.add(keras.layers.MaxPooling2D())
            
    
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(512,activation='relu'))
    model.add(keras.layers.Dense(CLASSES, activation='softmax'))
    return model
```

```python
'''
Base learner generation function
default base learner number is 10
'''
def generateBaseLearners(baseLearners = 10):
    #Random block and layer numbers
    blocks = np.random.randint(5, size= baseLearners)
    nConv = np.random.randint(5, size= baseLearners)
    models = []
    for i in range(baseLearners):
        name = "model{}".format(i+1)
        temp = modelFactory(blocks[i], nConv[i])
        models.append([temp, name])
    return models
        
```
