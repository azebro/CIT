```python
#Load Data 
def loadDataH5():
    with h5py.File('data/data1.h5','r') as hf: 
        trainX = np.array(hf.get('trainX')) 
        trainY = np.array(hf.get('trainY')) 
        valX = np.array(hf.get('valX')) 
        valY = np.array(hf.get('valY')) 
        print (trainX.shape,trainY.shape) 
        print (valX.shape,valY.shape)
    return trainX, trainY, valX, valY
trainX, trainY, testX, testY = loadDataH5()
```

```python
#Function to visualise the output from model run
def visualiseOutput(history):
    
    print("\n")
    print("Min Val Loss: ", min(history.history['val_loss']))
    print("Max Val Accuracy: ", max(history.history['val_accuracy']))
    print("\n")
    
    
    plt.figure(figsize=(20,10))
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
   
    plt.title('Training Loss and Accuracy')
    plt.ylabel('Loss/Accuracy')
    plt.xlabel('Epoch #')
    plt.legend(['train loss', 'val loss', 'train acc', 'val acc'], loc='upper right')
    plt.show()
```

```python
'''
Main model execution function
It takes the following parameters:
    1. a model
    2. an augmentation generator
    3. optional model name to name the checkpoint files
It will fit the model and checkpoint the improved weights to the file
This function will produce 1 weights file per model, if multiple are required the fname varaible needs
to be swapped for the commented one
'''
def executeAugmentedModel(model, generator, modelName = "model1"):
    #fname ="outputs/" + modelName + "_weights.{epoch:02d}-{val_loss:.2f}-{val_accuracy:.3f}.hdf5"
    fname ="outputs/" + modelName + "_weights.hdf5"
    checkpoint = tf.keras.callbacks.ModelCheckpoint(fname, monitor="val_accuracy", mode="max", save_best_only=True, 
                                                    verbose=0)
    trainGenerator = generator.flow(trainX, trainY, batch_size= BATCH_SIZE)
    h = model.fit(trainGenerator, 
                        validation_data=(testX, testY),  
                        epochs = NUM_EPOCHS, verbose=0, callbacks=[checkpoint])
    visualiseOutput(h)
    return min(h.history['val_loss']), max(h.history['val_accuracy'])
```

```python
''' 
Base learner trainign function, it takes collection of models
and executes them, capturing stats
'''
def trainBaseLearners(models):
    learnerStats = []
    for model in models:
        name = model[1]
        print("Training model: {}. \n\n".format(name))
        l, a = executeAugmentedModel(model[0], trainDataGenerator, name)
        data = [name, l, a]
        learnerStats.append(data)
    return learnerStats
    
```

```python
'''
Helper function to load previously saved model weights
This function assumes that the models are saved in 'outputs' folder
and each file is named with the following convention:
    modelName_weights.hdf5
'''
def loadModels(models):
    loadedModels = []
    for model in models:
        name = model[1]
        path = "outputs/{}_weights.hdf5".format(name)
        m = model[0]
        m.load_weights(path)
        loadedModels.append(m)
   
    return loadedModels
```

```python
'''
Function calculating ensemble average scores
It takes list of models, and validation data and 
preforms prediction followed by averaging the results
'''
def ensembleModels(models, testX):
    # predcict each model, collect outputs of models in a list
    y = np.array([model.predict(testX) for model in models])
    #Average out
    result = np.argmax(np.average(y, axis=0), axis=1)
    return result
```

```python
#Helper function to visualise the ensamble
#Referenced from: Matplotlib documentation

def visualiseEnsemble(data):
    labels = data[:,0]
    # the label locations
    x = np.arange(len(labels)) 
    # the width of the bars
    width = 0.55  

    fig, ax = plt.subplots(figsize=(10,5))
    rects1 = ax.bar(x , data[:,2].astype(float), width)

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Accuracy')
    ax.set_title('Base Learners and Ensemble Accuracy')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)

    def autolabel(rects):
        #Attach a text label above each bar in *rects*, displaying its height.
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{:.4f}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(rects1)
    fig.tight_layout()
    plt.show()
```

```python

WIDTH = trainX.shape[2]
HEIGHT = trainX.shape[1]
DEPTH = trainX.shape[3]
CLASSES = 17
NUM_EPOCHS = 200
BATCH_SIZE = 32
learning_rate = 0.0001
```

```python
inputShape = (HEIGHT, WIDTH, DEPTH)
includeTop = False
```

```python
#Mobilenet test 

trainDataGenerator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range= 45,
    zoom_range=0.3,
    width_shift_range=0.2, 
    height_shift_range=0.2,
    horizontal_flip=True
      )

trainGenerator = trainDataGenerator.flow(trainX, trainY, batch_size= BATCH_SIZE)
base_model = keras.applications.mobilenet.MobileNet(input_shape=inputShape, include_top=False, weights='imagenet')

#print("Number of layers in the base model: ", len(base_model.layers))
#base_model.summary()

#Enable re-train of the base model
base_model.trainable = True
# Fine tune from this layer onwards
fineTuneAt = len(base_model.layers) - 8

# Freeze all the layers before the `fineTuneAt` layer
for layer in base_model.layers[:fineTuneAt]:
    layer.trainable =  False
    
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    #Add 2 more dense layers
    keras.layers.Dense(1000,activation='relu'),
    keras.layers.Dense(1000,activation='relu'),
    keras.layers.Dense(CLASSES, activation='softmax')
])

#model.summary()
#Check different optimisers
opt = keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=True)
#opt = keras.optimizers.RMSprop(learning_rate=0.0001)
#opt = keras.optimizers.SGD(lr=0.01)
model.compile(loss="sparse_categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
h = model.fit(trainGenerator, validation_data=(testX, testY), epochs = 100, verbose=0)

visualiseOutput(h)

```

After selecting best tuned model, I will build a 5 base learner ensemble to see if accuracy can be brought even higher

```python
#The best performing model
def baseModel():
    base_model = keras.applications.VGG16(input_shape=inputShape, include_top=False, weights='imagenet')

    #print("Number of layers in the base model: ", len(base_model.layers))
    #base_model.summary()

    base_model.trainable = True
    # Fine-tune from this layer onwards
    fineTuneAt = len(base_model.layers) - 8

    # Freeze all the layers before the `fineTuneAt` layer
    for layer in base_model.layers[:fineTuneAt]:
        layer.trainable =  False

    model = tf.keras.Sequential([
        base_model,
        tf.keras.layers.GlobalAveragePooling2D(),
        #keras.layers.Flatten(),
        keras.layers.Dense(4906,activation='relu'),
        keras.layers.Dense(1000,activation='relu'),
        keras.layers.Dense(1000,activation='relu'),
        keras.layers.Dense(CLASSES, activation='softmax')
    ])

    #model.summary()
    opt = keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=False)
    #opt = keras.optimizers.RMSprop(learning_rate=0.0001)
    #opt = keras.optimizers.SGD(lr=0.01)
    model.compile(loss="sparse_categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
    return model
```

```python
'''
Ensemble  execution:
'''
#Define models, I will use the same base learner 5 times
models = [[baseModel(), "m1"], [baseModel(), "m2"], [baseModel(), "m3"], [baseModel(), "m4"], 
          [baseModel(), "m5"]]

#Train the base learners
stats = trainBaseLearners(models)

#Load saved weights
mods = loadModels(models)

#Score the ensemble
ensembleScore = accuracy_score(ensembleModels(mods, testX), testY)
stats.append(["ensemble", 0.0, ensembleScore])

#Visualise the output
visualiseEnsemble(np.array(stats))
```
